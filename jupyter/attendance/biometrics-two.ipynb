```python
import pandas as pd
import os

# Load the data
file_path = 'attendance-csv.csv'  # Replace with your file path
data = pd.read_csv(file_path, header=None)

# Assign dynamic column names based on structure
expected_columns = [
    'ID_Label', 'ID', 'Name_Label', 'Name', 'Dept_Label', 'Dept',
    'Day_16', 'Day_17', 'Day_18', 'Day_19', 'Day_20', 'Day_21', 'Day_22'
]

if len(data.columns) <= len(expected_columns):
    data.columns = expected_columns[:len(data.columns)]

# Clean the data
data = data.dropna(how='all', axis=1).dropna(how='all', axis=0)

# Check and clean 'Name' column
if 'Name' not in data.columns:
    raise KeyError("The 'Name' column is missing. Check the file format.")
attendance_columns = [col for col in data.columns if col.startswith('Day_')]

data['Days Present'] = data.apply(
    lambda row: sum(pd.to_datetime(row[attendance_columns], errors='coerce').notna()), axis=1
)

# Create summary
attendance_summary = data[['Name', 'Days Present']]

# Save the file to an accessible location
os.makedirs('output', exist_ok=True)
output_path = 'output/attendance_summary.csv'
attendance_summary.to_csv(output_path, index=False)
print(f"Summary saved at: {output_path}")
```
